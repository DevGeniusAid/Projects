## Financial Budget Analysis with Python

Each country has a financial budget that describes the government’s spending capacity in different sectors of the economy. In this article, I will walk you through the task of financial budget analysis with Python.

**What is a Financial Budget?**

There are so many Data Analysts today that come from a non-coding background. If you are from a commerce background then you may know what is a financial budget. In short, it is a detailed report on the income and expenditure of the government for a financial year.

You may get the task of analyzing a country’s financial budget every year if you are working as a data analyst in the media and communications field, as the media have to explain the government’s priorities for the complete financial year. 

**Import Libraries:**

```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

**Load dataset:**

```python
'''
Convert the dataset first
'''
url = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv'
response = requests.get(url)

with open('India_budget_2021.csv', 'wb') as file:
    file.write(response.content)
    
print('Succesful')

#Load data
raw_data = pd.read_csv('India_budget_2021.csv')
raw_data.head()
```

**Preprocess:**<br>
Create a copy/checkpoint:

```python
data_preprocessed = raw_data.copy()
pd.options.display.max_columns=None
pd.options.display.max_rows=None
display(data_preprocessed)
```

I can see a NaN value in this dataset, let’s remove the NaN values and continue with the task of financial budget analysis with Python:

```python
data_preprocessed.dropna()
```

I can see that not all the departments that are covered in this dataset are the main departments, as some departments can be covered in the others category. So let’s prepare the data by only selecting the main departments and putting all the other departments in the other category:

```python
data_preprocessed = data_preprocessed.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
data_to_append = pd.DataFrame([row])

# Resetting indices of data_preprocessed and data_to_append
data_preprocessed.reset_index(drop=True, inplace=True)
data_to_append.reset_index(drop=True, inplace=True)

data_preprocessed = pd.concat([data_preprocessed, pd.DataFrame([row])], ignore_index=True)
data_preprocessed
```

**Visualizatio:**

```python
'''
Now let’s plot this data to have a look at the priorities of the government for the financial year:
'''
# Create a bar plot with rotated x-axis labels
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.barplot(data=data_preprocessed, x='Department /Ministry', y='Fund allotted(in ₹crores)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

**Answer:**<br>
We can see that the finance department is getting the most of the share from the total budget of the government.

```python
'''
Now let’s plot this data into a donut plot to have a clear view of the distribution of funds among all the departments:
'''
# Extracting relevant data for the pie chart
budget_allocation = data_preprocessed["Fund allotted(in ₹crores)"]
departments = data_preprocessed["Department /Ministry"]

# Define custom colors for each wedge
custom_colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen', 'orange', 'lightpink', 'lightblue', 'lightgrey', 'cyan']

# Plotting a pie chart to visualize the distribution of the budget
plt.figure(figsize=(7, 7))
plt.pie(
    budget_allocation,
    labels=departments,
    autopct='%1.1f%%',  # Display percentage on each wedge
    startangle=90,  # Start the first wedge at 90 degrees (top of the circle)
    pctdistance=0.85,  # Adjust distance of percentage labels from the center
    shadow=True,  # Add shadow for better visual appeal
    #colors=custom_colors  # Use custom colors
)

# Adding a central circle to enhance the aesthetics
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)

# Adjusting font size and adding title
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)

# Display the pie chart
plt.show()
```

## Conclusion
We can see that the finance department is getting 40% of the funds. So this is how we can analyze a dataset that contains data about the revenue and expenditure of the government for a financial year.
